```r
# 2 - Do something with the data
# Calculate average mpg per carburetor 

AvgMpg = aggregate(mtcars$mpg~mtcars$carb, FUN=mean)
AvgMpg
```

```r
#3 - Export the data to CSV

write.csv(AvgMpg,file='AvgMpg.csv')
```
